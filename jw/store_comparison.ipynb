```python
sample = pd.read_csv('../data/01. final_sample', index_col = 0)
```

```python
store_month = sample.pivot_table(index=['year','month'],columns='store_nbr',values='units')
```

```python
store_month
```

```python
first = store_month[1]
first_2012 = first.loc[2012]
first_2013 = first.loc[2013]
first_2014 = first.loc[2014]
```

```python
plt.figure(figsize=(12,8))
plt.plot(first_2012,label='2012',c='blue',lw=4)
plt.plot(first_2013,label='2013',c='green',lw=4)
plt.plot(first_2014,label='2014',c='red',lw=4)
plt.legend(loc='best',prop={'size':20})
plt.show()
```
